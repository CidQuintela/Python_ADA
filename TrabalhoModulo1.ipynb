```python
def calculaDistancia(lista1 , pessoa):
  '''
  Lista1 -> Lista já classificada
  Pessoa -> item da lista não classificada
  '''
  lista = []
  listat = {}
  for i in range(len(lista1)):
    x = lista1[i][2]
    y = pessoa[2]
    #formula da distancia euclidiana, usando arredondamento para trabalhar com numeros menores
    distancia_xy = round((((x[0]-y[0])**2) + ((x[1]-y[1])**2) + ((x[2]-y[2])**2) + ((x[3]-y[3])**2)) **(1/2))
    classif = [distancia_xy,lista1[i][1]]

    lista.append(classif)
  lista.sort()

  return lista
```

```python
def retornaK (lista,k):
  '''
  lista -> output da função calculaDistancia
  '''
  first_k_qtd = lista[:k]
  listaClassificacao = []
  for i in range(len(first_k_qtd)):
    classificacao = first_k_qtd[i][1]
    listaClassificacao.append(classificacao)

  return listaClassificacao
```

```python
def verificaClassificacao (listaClassif):
  '''
  listaClassif -> output da função retornaK
  '''
  classificacao = ""
  promotor = listaClassif.count("Promotor")
  neutro = listaClassif.count("Neutro")
  detrator = listaClassif.count("Detrator")

  if(promotor > neutro and promotor > detrator):
    classificacao = "Promotor"
  elif(neutro > promotor and neutro > detrator):
    classificacao = "Neutro"
  elif(detrator > neutro and detrator > promotor):
    classificacao = "Detrator"

  return classificacao
```
